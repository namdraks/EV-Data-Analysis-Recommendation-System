```python
# Rename all columns: remove extra spaces and standardize
df.columns = df.columns.str.strip()
```

```python
# Confirm and clean the correct column
df['mean - Energy consumption [kWh/100 km]'] = pd.to_numeric(df['mean - Energy consumption [kWh/100 km]'], errors='coerce')

# Calculate Q1 and Q3
Q1 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.25)
Q3 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
# Find the EVs that are outliers
outliers = df[
    (df['mean - Energy consumption [kWh/100 km]'] < lower_bound) |
    (df['mean - Energy consumption [kWh/100 km]'] > upper_bound)
]

# Display the outlier vehicles
outliers
```

```python
print(f"Number of Outlier Vehicles: {outliers.shape[0]}")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['mean - Energy consumption [kWh/100 km]'], color='lightblue')
plt.title('Energy Consumption Outliers (kWh/100km)', fontsize=16)
plt.grid()
plt.show()
```

```python
# Applying the IQR method to detect outliers in the "mean - energy consumption [kWh/100 km]" column, we observed that there were no outliers present in the dataset.
# This indicates that the energy consumption among the electric vehicles is highly consistent.
# Additionally, the boxplot visualization confirmed that all data points fall within the normal statistical range, showing no signs of unusually inefficient models.
```

# Task 4: Build an EV recommendation class.

```python
# Define the EVRecommender class
class EVRecommender:
    def __init__(self, data):
        self.data = data

    # Method to collect user preferences for budget, driving range, and battery capacity
    def get_user_input(self):
        self.budget = float(input("Enter your budget (PLN): "))
        self.min_range = float(input("Enter your desired minimum range (km): "))
        self.min_battery_capacity = float(input("Enter your minimum battery capacity (kWh): "))
        print(f"User Preferences — Budget: {self.budget} PLN, Minimum Range: {self.min_range} km, Battery Capacity: {self.min_battery_capacity} kWh")

    # Method to filter and recommend EVs based on user preferences
    def recommend(self):
        filtered_evs = self.data[
            (self.data['Minimal price (gross) [PLN]'] <= self.budget) &
            (self.data['Range (WLTP) [km]'] >= self.min_range) &
            (self.data['Battery capacity [kWh]'] >= self.min_battery_capacity)
        ]
        
        # Return the top 3 EVs sorted by range
        return filtered_evs[['Car full name', 'Make', 'Minimal price (gross) [PLN]', 'Range (WLTP) [km]', 'Battery capacity [kWh]']]\
            .sort_values(by='Range (WLTP) [km]', ascending=False)\
            .head(3)

# Instantiate the recommender with the dataset
recommender = EVRecommender(df)

# Collect user input
recommender.get_user_input()

# Generate and display the top 3 EV recommendations
top_3_evs = recommender.recommend()
top_3_evs
```

```python
# This class, EVRecommender, is designed to allow users to input their budget, desired range, and minimum battery capacity.
# The get_user_input method collects these details interactively from the user.
# The recommend method filters the dataset based on the user’s input and returns the top 3 EVs sorted by the driving range.
```

# Task 5: Inferential Statistics – Hypothesis Testing

```python
from scipy.stats import ttest_ind
```

```python
# Filter Tesla and Audi data
tesla_engine_power = df[df['Make'] == 'Tesla']['Engine power [KM]']
audi_engine_power = df[df['Make'] == 'Audi']['Engine power [KM]']

# Check how many samples we have
print(f"Number of Tesla samples: {tesla_engine_power.shape[0]}")
print(f"Number of Audi samples: {audi_engine_power.shape[0]}")
```

```python
# Perform two-sample t-test
t_stat, p_value = ttest_ind(tesla_engine_power, audi_engine_power, equal_var=False)  # Welch's t-test safer if variance not equal

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Interpret the result
if p_value < 0.05:
    print("There is a significant difference in Engine Power between Tesla and Audi.")
else:
    print("No significant difference in Engine Power between Tesla and Audi.")
```

```python
# We conducted a two-sample t-test to determine whether there is a significant difference in the average engine power between Tesla and Audi vehicles.
# As the p-value was found to be below 0.05, we can conclude that the engine power of Tesla and Audi electric vehicles differs significantly.
```
